# Lab 10: Tic-Tac-Toe (solutions)

**Note:** These solutions use a simpler approach to print the board.

```python
# Make a 2D list of all zeros with given dimensions.
def make2dList(rows, cols):
    g = []
    for row in range(0, rows): 
        g.append([0] * cols)
    return g

# Print the TTT board to the screen using X's and O's.
# Use the standard nested loops.
def print_board(board):
    for row in range(len(board)):
        for col in range(len(board[row])):
            if board[row][col] < 0:
                print("O", end="")
            elif board[row][col] > 0:
                print("X", end="")
            else:
                print("-", end="")
        print()

# Add up and return the numbers in row r of the board.
# (Can do with a loop or manually.)
def add_row(board, r):
    return sum(board[r])
    
# Add up and return the numbers in column c of the board.
# (Can do with a loop or manually.)
def add_column(board, c):
    return board[0][c] + board[1][c] + board[2][c]
    
# Add up and return the numbers on the upper-left to
# lower-right of the board. (Can do with a loop or manually.)
def add_diagonal1(board):
    return board[0][0] + board[1][1] + board[2][2]

# Add up and return the numbers on the upper-right to
# lower-left of the board. (Can do with a loop or manually.)
def add_diagonal2(board):
    return board[0][2] + board[1][1] + board[2][0]
    
# Check if all the squares on the board are full (non-zero).
# Return True if they're all full, and False if they're not.
def board_full(board):
    return not (0 in board[0] or 0 in board[1] or 0 in board[2])

# Check to see if the sum of any row, column, or diagonal is num.
def any_sum(board, num):
    return add_row(board, 0) == num \
        or add_row(board, 1) == num \
        or add_row(board, 2) == num \
        or add_column(board, 0) == num \
        or add_column(board, 1) == num \
        or add_column(board, 2) == num \
        or add_diagonal1(board) == num \
        or add_diagonal2(board) == num

# Check to see if the sum of any row, column, or diagonal is 3.
# Return True for an X win, and False otherwise.
def x_wins(board):
    return any_sum(board, 3)
    
# Check to see if the sum of any row or column is -3.
# Also check the diagonals for a win for O.
def o_wins(board):
    return any_sum(board, -3)

def main():
    board = make2dList(3, 3)
    current_player = "X"

    while not x_wins(board) and not o_wins(board) and not board_full(board):
        
        print_board(board)
        print()

        while True:
            print("Enter move for player ", current_player, ".", sep="")
            row = int(input("Row (0-2): "))
            col = int(input("Column (0-2): "))
            
            if row < 0 or row > 2:
                print("Invalid row.")
            elif col < 0 or col > 2:
                print("Invalid column.")
            elif board[row][col] != 0:
                print("Square already taken.")
            else:
                break
                
            print()
        
        print()

        # Place the correct piece on the board.
        if current_player == "X":
            board[row][col] = 1
        else:
            board[row][col] = -1

        # Switch the player.
        if current_player == "X":
            current_player = "O"
        else:
            current_player = "X"

    # Loop is over, therefore the game has ended.
    print_board(board)
    print()
    
    if x_wins(board):
        print("X wins.")
    elif o_wins(board):
        print("Y wins.")
    else:
        print("Tie.")
    
main()
```
